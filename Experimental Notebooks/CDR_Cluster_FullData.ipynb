```
import pandas as pd
```

```
data = pd.read_csv("https://raw.githubusercontent.com/equester/TataTelecomAI/master/Sample%20Data/cdrfullsample.csv")
```

```
data.head(10)
```

```
def SFTAg(eachValue):
  if eachValue == "*** NO ANSWER ***":
    return 1
  else:
    return 0
```

```
data['SAFlag'] = data['Answer'].apply(SFTAg)
```

```
data.head(10)
```

```
data = data.drop(['Answer'], axis=1)
```

```
data.head()
```

```
df = pd.concat([data.drop('Completion Code', axis=1), pd.get_dummies(data['Completion Code']),pd.get_dummies(data['Groupby5mins'])], axis=1)
```

```
df.head()
```

```
df = df.drop(['Groupby5mins'], axis =1)
```

```
df.head(10)
```

```
from sklearn.cluster import KMeans
model=KMeans(n_clusters= 4)
model.fit(df)
clusassign = model.predict(df)
df['Cluster'] = clusassign
```

```
df.head()
```

```
df.to_csv("clusterdata_cdrfulldata.csv")
```

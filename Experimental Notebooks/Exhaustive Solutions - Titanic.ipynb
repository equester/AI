# Get Data 

```
train_link = 'https://raw.githubusercontent.com/pcsanwald/kaggle-titanic/master/train.csv'
test_link = 'https://raw.githubusercontent.com/pcsanwald/kaggle-titanic/master/test.csv'
```

## Pandas Read 

```
start = time.time()
pd_train = pd.read_csv(train_link)
pd_test = pd.read_csv(test_link)
end = time.time()
print(end - start)
```

## DataTable Read 

```
start = time.time()
dt_train  = dt.fread(train_link)
dt_test = dt.fread(test_link)
end = time.time()
print(end - start)
```

## Dask Read 

```
# Dask seems to be not working for direct csv links. Uploaded csv files! 
start = time.time()
dask_train = dd.read_csv('train.csv')
dask_test = dd.read_csv('test.csv')
end = time.time()
print(end - start)
```

## Vaex Read 

```
start = time.time()
vx_train = vx.read_csv(train_link)
vx_test = vx.read_csv(test_link)
end = time.time()
print(end - start)
```

```
import pandas as pd
```

```
%%time
data = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Musical_Instruments_5.json.gz", lines=True)
from gensim.parsing.preprocessing import remove_stopwords
doc_final = remove_stopwords("-".join(data['reviewText'].tolist()))
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english')) 
  
word_tokens = word_tokenize(doc_final) 
  
filtered_sentence = [w for w in word_tokens if w not in stop_words] 
import gensim
%time documents = gensim.utils.simple_preprocess(" ".join(filtered_sentence))
model = gensim.models.Word2Vec(
        [documents],
        size=300,
        window=20,
        min_count=2,
        workers=20,
        iter=10)
  
```

```
filtered_sentence
```

```
word = ["star wars"]
model.wv.most_similar(positive=word)
```

```
model.similarity('funny','stories')
```

```
vector1 = model['cheap']+model['book']+model['kindle']
vector2 = model['star']+model['war']+model['battle']

model.similarity(vector1,vector2)
```

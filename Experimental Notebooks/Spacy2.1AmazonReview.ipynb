```
nlp = spacy.load('enmd')
# % time data_s['reviewText'] = data_s.reviewText.parallel_apply(nlp)
% time data_s['reviewText'] = data_s.reviewText.apply(nlp)
```

```
def extract_SVO(x):
  tuples_list = [] 
  tuples = textacy.extract.subject_verb_object_triples(x)
  if tuples:
    tuples_to_list = list(tuples)
    tuples_list.append(tuples_to_list)
  return tuples_list
```

```
def getSGRankTopics(eachValue):
#     text = nlp(eachValue)
    tuples_list = [] 
    tuples = textacy.keyterms.sgrank(eachValue, ngrams=(1, 2, 3, 4, 5, 6), normalize='lemma', window_width=1500, n_keyterms=10, idf=None)
    if tuples:
      tuples_to_list = list(tuples)
      tuples_list.append(tuples_to_list)
    return tuples_list
```

```
%time data_s['SGRank'] = data_s.reviewText.apply(getSGRankTopics)
```

```
%time data_s['SVOPair'] = data_s.reviewText.apply(extract_SVO)
```

```
raw_text_raw = data_review['reviewText'][1:11]
raw_text = ".".join(data_review['reviewText'][1:11])
text = nlp(raw_text)
print(text)
print(list(textacy.extract.subject_verb_object_triples(text)))
print(list(textacy.keyterms.key_terms_from_semantic_network(text, normalize='lemma', window_width=2, edge_weighting='binary', ranking_algo='pagerank', join_key_words=False, n_keyterms=5)))
print(list(textacy.keyterms.sgrank(text, ngrams=(1, 2, 3, 4, 5, 6), normalize='lemma', window_width=1500, n_keyterms=10, idf=None)))
print(list(textacy.keyterms.textrank(text, normalize='lemma', n_keyterms=10)))

from nltk.corpus import stopwords
noun_phrase_pairs = []
for chunk in text.noun_chunks:
#   print (chunk.text)
  if len(chunk.text.split(" "))>1 and chunk.text not in nltk.corpus.stopwords.words('english'):
    noun_phrase_pairs.append(chunk.text)
print (noun_phrase_pairs)


noun_adj_pairs = []
for i,token in enumerate(text):
    if token.pos_ not in ('NOUN','PROPN','VERB'):
        continue
    for j in range(i+1,len(text)):
        if text[j].pos_ == 'ADJ':
            noun_adj_pairs.append((token.text,text[j].text))
            break
print (noun_adj_pairs)

r = Rake()
# Extraction given the list of strings where each string is a sentence.
r.extract_keywords_from_text(raw_text)
# To get keyword phrases ranked highest to lowest with scores.
print(r.get_ranked_phrases_with_scores())



negation_tokens = [tok for tok in text if tok.dep_ == 'neg']
negation_head_tokens = [token.head for token in negation_tokens]
for token in negation_head_tokens:
    print(token.text, token.dep_, token.head.text, token.head.pos_, [child for child in token.children])

    
displacy.render(text, style="ent",jupyter=True)


from textblob import TextBlob
TextBlob(raw_text).sentiment


from textacy.vsm import Vectorizer
terms_list=[[tok  for tok in doc.split() if tok not in stopwords.words('english') ] for doc in raw_text_raw]
vectorizer = Vectorizer(tf_type='linear', apply_idf=True, idf_type='smooth',min_df=2, max_df=0.95)
doc_term_matrix = vectorizer.fit_transform(terms_list)
model = textacy.tm.TopicModel('nmf', n_topics=5)
model.fit(doc_term_matrix)
doc_topic_matrix = model.transform(doc_term_matrix)

for topic_idx, top_terms in model.top_topic_terms(vectorizer.id_to_term):
  print('topic', topic_idx, ':', '   '.join(top_terms))
model.termite_plot(doc_term_matrix, vectorizer.id_to_term, topics=-1,  n_terms=25, sort_terms_by='seriation') 
```

```
# Genral Code 

df = data_subset.merge(tag, on= "name")
```

```
TID_Groupby = df.groupby("name")["tid"].count().reset_index(name='count')
```

```
TID_Groupby.head()
```

```
TID_Groupby['name'] = TID_Groupby.index
```

```
TID_Groupby[(TID_Groupby['name']=='Awareness')]['count'].values[0]
```

```
def getTIDNumber(eachValue):
  return TID_Groupby[(TID_Groupby['name']==eachValue)]['count'].values[0]
```

```
data_subset['Total TID'] = data_subset['name'].apply(getTIDNumber)
```

```
data_subset['Non-English TID'] = data_subset['name'].apply(getNonEnglishTID)
```

```
def getEnglisALias(eachvalue):
  try:
    return df[(df['name']==eachvalue)][df['langcode_y'] == 'en']['alias_y'].values
  except:
    return "No Alias"
```

```
def getNonEnglisALias(eachvalue):
  try:
    return df[(df['name']==eachvalue)][df['langcode_y'] != 'en']['alias_y'].values
  except:
    return "No Alias"
```

```
def getTotalNumberofArticle(eachValue):
  try:
    return df3[df3['tag_id']==eachValue].shape[0]
  except:
    return "No Article Found"
```

```
def getAllArticlesNodeID(eachValue):
  try:
    return df3[df3['tag_id']==eachValue]['node_id'].tolist()
  except:
    return "No Node ID Found"
  
```

```
data_subset['EnglishAlias'] = data_subset['name'].apply(getEnglisALias)
```

```
data_subset['NonEnglishAlias'] = data_subset['name'].apply(getNonEnglisALias)
```

```
data_subset['TotalArticle'] = data_subset['Non-English TID'].apply(getTotalNumberofArticle)
```

```
data_subset['ArticleNode'] = data_subset['Non-English TID'].apply(getAllArticlesNodeID)
```

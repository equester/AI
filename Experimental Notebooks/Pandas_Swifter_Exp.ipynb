```
import pickle
with open('data.pickle', "rb") as f:
    data = pickle.load(f)
```

```
import pandas as pd
df = pd.DataFrame({'Translate': data})
```

```
df.head()
```

```
import swifter
from textblob import TextBlob

def GetSentiment(eachValue):
  try:
    return TextBlob(eachValue).sentiment
  except TypeError:
    return []

df['Sentiment'] = df['Translate'].swifter.apply(GetSentiment)
```

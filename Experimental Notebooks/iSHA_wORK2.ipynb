```
# Usecase 0 ----> Taggingg Working Not Working 
def getErrorCodeofLink(eachValue):
  try:
    print(requests.head('https://isha.sadhguru.org/in/en/wisdom/topic{}'.format(eachValue)).status_code)
    return requests.head('https://isha.sadhguru.org/in/en/wisdom/topic{}'.format(eachValue)).status_code
  except:
    return "No Alias"
  
data['StatusAutoCode'] = data['alias'].apply(getErrorCodeofLink)
data.to_csv("StatusCodeOutput.csv")
```

```
# Use_Case 1 -> Get Node ID of all that has non-english mapping og not working web-pages

data_subset = data[(data['To be looked']=='yes') & (data['Status'] == "n") | (data['Status'] == "N") ]

# Merge with Tag data to get all TAG related data and Other Language TAGID
df = data_subset.merge(tag, on= "name")
#merging artcile data
df3 = df.merge(article, left_on="tid", right_on="tag_id" )

TID_Groupby = df.groupby("name")["tid"].count().reset_index(name='count')
# print(TID_Groupby[(TID_Groupby['name']=='Adiyogi Alayam')]['count'].values[0])
# print(TID_Groupby.head())
# Column 1 -> Total TID of given NAME variable 
def getTIDNumber(eachValue):
  return TID_Groupby[(TID_Groupby['name']==eachValue)]['count'].values

data_subset['Total TID'] = data_subset['name'].apply(getTIDNumber)

# Column 2 -> Get non-english TID Number 
def getNonEnglishTID(eachValue):
  try:
    return df[(df['name']==eachValue)][df['langcode_y'] != 'en']['tid'].values
  except:
    return "No Non-ENglish TID"
  
data_subset['Non-English TID'] = data_subset['name'].apply(getNonEnglishTID)

# Column 3,4,5,6 ->  
def getEnglisALias(eachvalue):
  try:
    return df[(df['name']==eachvalue)][df['langcode_y'] == 'en']['alias_y'].values
  except:
    return "No Alias"
  
def getNonEnglisALias(eachvalue):
  try:
    return df[(df['name']==eachvalue)][df['langcode_y'] != 'en']['alias_y'].values
  except:
    return "No Alias"
  
def getTotalNumberofArticle(eachValue):
  try:
    if len(eachValue) == 1:
      return df3[df3['tag_id']==eachValue[0]].shape[0]
    else:
      return df3[df3['tag_id']==eachValue[0]].shape[0] + df3[df3['tag_id']==eachValue[1]].shape[0]
  except:
    return "No Article Found"
  

def getAllArticlesNodeID(eachValue):
  try:
    return df3[df3['tag_id']==eachValue[0]]['node_id'].tolist()
  except:
    return "No Node ID Found"
  

data_subset['EnglishAlias'] = data_subset['name'].apply(getEnglisALias)
data_subset['NonEnglishAlias'] = data_subset['name'].apply(getNonEnglisALias)
data_subset['TotalArticle'] = data_subset['Non-English TID'].apply(getTotalNumberofArticle)
data_subset['ArticleNode'] = data_subset['Non-English TID'].apply(getAllArticlesNodeID)
```

```

from fastText_multilingual.fasttext import FastVector
en_dictionary = FastVector(vector_file='wiki.multi.en.vec')
id_dictionary = FastVector(vector_file='wiki.multi.id.vec')

```

```
eng_word = "love"
id_word = "cinta"

en_vector = en_dictionary[eng_word]
id_vector = id_dictionary[id_word]
print(FastVector.cosine_similarity(en_vector, id_vector))

en_dictionary_bk = en_dictionary
id_dictionary_bk = id_dictionary

en_dictionary_bk.apply_transform('fastText_multilingual/alignment_matrices/en.txt')
id_dictionary_bk.apply_transform('fastText_multilingual/alignment_matrices/id.txt')

en_vector_n = en_dictionary_bk[eng_word]
id_vector_n = id_dictionary_bk[id_word]
print(FastVector.cosine_similarity(en_vector_n, id_vector_n))
```

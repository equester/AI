```
import pandas as pd
import spacy
```

```
data = pd.read_csv("TopicDFFinal.csv")
```

```
data.columns
```

```
def getNLPObject(eachValue):
  return nlp(eachValue)

def getNoun(eachValue):
  Noun = []
  for token in eachValue:
    if token.pos_ == "NOUN":
      Noun.append(token.text)
  return Noun

def getAdjective(eachValue):
  ADJ = []
  for token in eachValue:
    if token.pos_ == "ADJ":
      ADJ.append(token.text)
  return ADJ
  
def getVerb(eachValue):
  Verb = []
  for token in eachValue:
    if token.pos_ == "VERB":
      Verb.append(token.text)
  return Verb

def getAdVerb(eachValue):
  ADV = []
  for token in eachValue:
    if token.pos_ == "ADV":
      ADV.append(token.text)
  return ADV
```

```
data_s = data[:200]
```

```
% time
nlp = spacy.load("enmd")
% time
data['SpacyDoc'] = data.ImportantTopic.apply(getNLPObject)
```

```
len(data)
```

```
%time
data['Noun'] = data.SpacyDoc.apply(getNoun)
data['Adjective'] = data.SpacyDoc.apply(getAdjective)
data['Verb'] = data.SpacyDoc.apply(getVerb)
data['Adverb'] = data.SpacyDoc.apply(getAdVerb)
```

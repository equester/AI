```
import pandas as pd
```

```
data = pd.read_csv("pepredict.csv")
```

```
data.head()
```

```
columns_totake = ['PE_Monthly','Nav_Date','Debt_CurrentV','Equity_CurrentV','Class']
data_s = data[columns_totake]
```

```
data_s['Class'][205].split()[0]
```

```
type(data_s['Class'][0])
```

```
data_s.info()
```

```
data_s['Class'] = data_s['Class'].str.replace('-','\t0')
```

```
data_s['Class'][205].split()[0]
```

```
def cleanClass(eachValue):
  return eachValue.split()[0]
  
```

```
data_s['cleanclass'] = data_s['Class'].apply(cleanClass)
```

```
data_s.columns
```

```
data_s.drop("Class",inplace=True,axis=1)
```

```
data_s['Nav_Date']= pd.to_datetime(data_s['Nav_Date']) 
```

```
data_s['year'] = pd.DatetimeIndex(data_s['Nav_Date']).year
data_s['month'] = pd.DatetimeIndex(data_s['Nav_Date']).month
```

```
data_s.drop("Nav_Date",axis=1,inplace=True)
```

```
X = data_s[['PE_Monthly', 	'year'	,'month']]
y = data_s['cleanclass']
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_Scaled = scaler.transform(X)
```

```
from tpot import TPOTRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.80, test_size=0.20)

tpot = TPOTRegressor(generations=50, population_size=50, verbosity=2)
tpot.fit(X_train, y_train)
```

```
print(tpot.score(X_test, y_test))
```

```
print(tpot.fitted_pipeline_)
```

```
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.80, test_size=0.20)

exported_pipeline = tpot.fitted_pipeline_.fit(X_train, y_train)
results = exported_pipeline.predict(X_test)
from sklearn import metrics 
metrics.median_absolute_error(y_test,results)
```

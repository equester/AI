```
import yake

text_content = """
	Nxkh̄enụ̄x cāk phækh kec h̄lām thī̀ xṭhibāy wị̂ thī̀ nī̀ reā yạng h̄ı̂kār s̄āṭhit læa api"""
  
# assuming default parameters
simple_kwextractor = yake.KeywordExtractor()
keywords = simple_kwextractor.extract_keywords(text_content)

for kw in keywords:
	print(kw)

# specifying parameters
custom_kwextractor = yake.KeywordExtractor(lan="th", n=3, dedupLim=0.9, dedupFunc='seqm', windowsSize=1, top=20, features=None)

keywords = custom_kwextractor.extract_keywords(text_content)

for kw in keywords:
	print(kw)
```

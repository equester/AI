```
import pandas as pd
data = pd.read_csv("g_translate_dump.csv")
from textblob import TextBlob
def getSnetiment(eachValue):
  return TextBlob(eachValue).sentiment.polarity

def sentClass(eachValue):
  if eachValue == 0.0:
    return "neutral"
  elif eachValue > 0.0:
    return "positive"
  else:
    return "negative"

data['sentiment'] = data['raw_text'].apply(getSnetiment)
data['class'] = data['sentiment'].apply(sentClass)
data['doc_text'] = data['translated_text']
df = data[['doc_text','class']]
df.to_csv("file.csv")
```

```
# creating single sentiment data
import pandas as pd

#Amazon Review Data
amazondata = pd.read_csv("/content/ludwig/file-3.csv")
# amazondata.head()

#Soco Data 
socodata = pd.read_csv("socodata.csv")
df = amazondata.append(socodata)
# socodata.drop(['Unnamed: 2','Unnamed: 3'],inplace=True)

# News Sentiment Data 
news_data = pd.read_csv("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/news-sentiment/sentiment-data-v2.csv")
news_data.columns = ['class','doc_text']
df = df.append(news_data)

# Adding Amazon, Yelp, IMDB

#Amazon
df_a = pd.read_json("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/multidomain-sentiment/bm-amazon.json",lines=True)
df_a['negative'][0]
df_a_neg = pd.DataFrame()
df_a_neg['doc_text'] = df_a['negative'][0]
df_a_neg['class'] = 'negative'
# df_a_neg.head()
df_a_pos = pd.DataFrame()
df_a_pos['doc_text'] = df_a['positive'][0]
df_a_pos['class'] = 'positive'
# df_a_pos.head()
df = df.append(df_a_pos)
df = df.append(df_a_neg)

#Yelp 
df_a = pd.read_json("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/multidomain-sentiment/bm-imdb.json",lines=True)
df_a['negative'][0]
df_a_neg = pd.DataFrame()
df_a_neg['doc_text'] = df_a['negative'][0]
df_a_neg['class'] = 'negative'
# df_a_neg.head()
df_a_pos = pd.DataFrame()
df_a_pos['doc_text'] = df_a['positive'][0]
df_a_pos['class'] = 'positive'
# df_a_pos.head()
df = df.append(df_a_pos)
df = df.append(df_a_neg)


#IMDB 
df_a = pd.read_json("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/multidomain-sentiment/bm-yelp.json",lines=True)
df_a['negative'][0]
df_a_neg = pd.DataFrame()
df_a_neg['doc_text'] = df_a['negative'][0]
df_a_neg['class'] = 'negative'
# df_a_neg.head()
df_a_pos = pd.DataFrame()
df_a_pos['doc_text'] = df_a['positive'][0]
df_a_pos['class'] = 'positive'
# df_a_pos.head()
df = df.append(df_a_pos)
df = df.append(df_a_neg)

# Twitter 

#positive
df_tw = pd.DataFrame()
for x in range(1,8):
  # print (x)
  if x == 1:
    # print ("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/twitter-sentiment/positive/%d.json" % x)
    df_tw = pd.read_json("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/twitter-sentiment/positive/%d.json" % x)
  else:
    df_tw.append(pd.read_json("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/twitter-sentiment/positive/%d.json" % x))
df_tw.columns = ['doc_text']
df_tw['class'] = 'positive'
df = df.append(df_tw)

#negative 
df_tw = pd.DataFrame()
for x in range(1,10):
  # print (x)
  if x == 1:
    # print ("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/twitter-sentiment/negative/%d.json" % x)
    df_tw = pd.read_json("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/twitter-sentiment/negative/%d.json" % x)
  else:
    try:
      # print ("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/twitter-sentiment/negative/%d.json" % x)
      df_tw.append(pd.read_json("https://raw.githubusercontent.com/huseinzol05/Malaya-Dataset/master/twitter-sentiment/negative/%d.json" % x))
    except:
      pass
df_tw.columns = ['doc_text']
df_tw['class'] = 'negative'
df = df.append(df_tw)
```

```
def getclassnamecorrect(eachValue):
  if eachValue == "Positive":
    return "positive"
  elif eachValue == "Negative":
    return "negative"
  else:
    return eachValue
df['class'] = df['class'].apply(getclassnamecorrect)
```

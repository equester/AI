# Acronym 

```
import base64
import requests

Accronym = []
url = 'https://raw.githubusercontent.com/ramaprakoso/analisis-sentimen/master/kamus/acronym.txt'

req = requests.get(url)
Accronym.extend(req.text.split("\n"))
AcronymDict = {}
for eachLine in req.text.split("\n"):
  AcronymDict[eachLine.split()[0]] = eachLine.split()[2]
  
```

#Twitter Shorthands 

```
import base64
import requests

TwitterShortHand = {}
url = 'https://raw.githubusercontent.com/ramaprakoso/analisis-sentimen/master/kamus/kbba.txt'
req = requests.get(url)

ShortHandDict= {}
for eachLine in req.text.split("\n"):
  ShortHandDict[eachLine.split("\t")[0]] = eachLine.split("\t")[1].split("\r")[0]
  
```

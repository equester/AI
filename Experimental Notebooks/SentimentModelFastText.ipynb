```
def sentClass(eachValue):
  if eachValue == 0.0:
    return "__label__neutral"
  elif eachValue > 0.0:
    return "__label__positive"
  else:
    return "__label__negative"
```

```
import fasttext
```

```
model = fasttext.train_supervised('ind.train',lr=1,dim=300,epoch=5,wordNgrams=2,loss="hs",thread=-1,pretrainedVectors="cc.id.300.vec",verbose=2)
```

```
def print_results(N, p, r):
    print("N\t" + str(N))
    print("P@{}\t{:.3f}".format(1, p))
    print("R@{}\t{:.3f}".format(1, r))

print_results(*model.test('ind.valid'))
```

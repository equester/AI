```
# Sentiment Block 

# Multimonial Model 

def getsentiment_multinomial(eachValue):
  return model_mm.predict(eachValue, get_proba=True)

def getMultinomialModel():
  return malaya.sentiment.multinomial()

# XGBoost Model 
def getsetiment_XGBoost(eachValue):
  return model_xgb.predict(eachValue,get_proba=True)

def getXGBoostModel():
  return malaya.sentiment.xgb()

# Attention Model 
def getsentiment_AttentionModel(eachValue):
  try:
    return model_at.predict(eachValue,get_proba=True)
  except ValueError:
    return ""

def getAttentionModel():
  return malaya.sentiment.deep_model('bahdanau')

# Bert Model 
def getsentiment_BertModel(eachValue):
  return model_b.predict(eachValue,get_proba=True)

def getBertModel():
  return malaya.sentiment.bert()

# Stacked Model 
def getsentiment_StackedModel(eachValue):
  return malaya.stack.predict_stack([model_mm, model_xgb, model_at], eachValue)

def getStackedModel():
  multi = getMultinomialModel()
  xgb = getXGBoostModel()
  attention = getAttentionModel()
  return multi,xgb,attention
```

```
import yake
# simple_kwextractor = yake.KeywordExtractor()
custom_kwextractor = yake.KeywordExtractor(lan="en", n=3, dedupLim=0.9, dedupFunc='seqm', windowsSize=1, top=5, features=None)


def getYake(eachValue):
  try:
    return custom_kwextractor.extract_keywords(eachValue)
  except ValueError:
    return None

df_sa_clean['TopicModel1'] = df_sa_clean[58].apply(getYake)
```

```
import textacy
import textacy.keyterms
```

```
def getsemanticTopic(eachValue):
  text = nlp(eachValue)
  return list(textacy.keyterms.key_terms_from_semantic_network(text, normalize='lemma', window_width=2, edge_weighting='binary', ranking_algo='pagerank', join_key_words=False, n_keyterms=5))

df_sa_clean['TopicModel2'] = df_sa_clean[58].apply(getsemanticTopic)
```

```
def getSGRankTopic(eachValue):
  text = nlp(eachValue)
  return list(textacy.keyterms.sgrank(text, ngrams=(1, 2, 3, 4, 5, 6), normalize='lemma', window_width=1500, n_keyterms=10, idf=None))

df_sa_clean['TopicModel3'] = df_sa_clean[58].apply(getSGRankTopic)
```

```
def getTextRankTopic(eachValue):
  text = nlp(eachValue)
  return list(textacy.keyterms.textrank(text, normalize='lemma', n_keyterms=10))

df_sa_clean['TopicModel4'] = df_sa_clean[58].apply(getTextRankTopic)
```

```
import pandas as pd
```

```
data = pd.read_csv("cc_data.csv")
```

```
data.head()
```

```
data.columns = ['ds','y']
```

```
from fbprophet import Prophet
```

```
import numpy as np
data = data.dropna()
```

```
m = Prophet()
m.fit(data)
```

```
future = m.make_future_dataframe(periods=5)
future.head()
```

```
# Python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```
m.plot(forecast)
```
